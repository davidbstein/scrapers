```python
def parse_maps_bundle(bundle):
    DAYS = ['sun', 'mon', 'tues', 'wed', 'thurs', 'fri', 'sat']
    d = dict()
    d['address'] = bundle[2], bundle[18]
    d['phone'] = bundle[3]
    d['expense'] = bundle[4][2]
    d['reviews'] = bundle[4][3][1]
    d['website'] = bundle[7][1]
    d['coords'] = bundle[9]
    d['name'] = bundle[11]
    d['categories'] = bundle[13]
    d['neighborhood'] = bundle[14]
    d['link'] = bundle[27]
    d['description'] = bundle[32][1][1]
    d['open_times'] = [e[:2] for e in bundle[34][1]]
    d['crowdedness'] = [[DAYS[e[0]-1], e[1]] for e in bundle[84][0]]
    d['features'] = [[e[1], [ee[1] for ee in e[2]]] for e in bundle[100][1]]
    return d
```
